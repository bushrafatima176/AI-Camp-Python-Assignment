```python
# Q7 Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it and generate mark sheet using grades ?

math = int(input("Enter Math Marks : "))
eng = int(input("Enter Eng Marks : "))
urdu = int(input("Enter Urdu Marks : "))
isl = int(input("Enter Isl Marks : "))
comp = int(input("Enter Comp Marks : "))
total = math+eng+urdu+isl+comp
per = total/500 * 100
print("Total marks of all subject out of 500:", total)
if(per>=80 and per<=100):
    print("Grade A+")
elif(per>=70 and per<80):
    print("Grade A")
elif(per>=60 and per<70):
    print("Grade B")
elif(per>=50 and per<60):
    print("Grade C")
elif(per>=40 and per<50):
    print("Grade D")
elif(per>=33 and per<40):
    print("Grade #")
else:
    print("Fail")
```
